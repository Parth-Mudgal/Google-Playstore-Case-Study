# Google Playstore Case Study

**Problem Statement**

The team at Google Play Store wants to develop a feature that would enable them to boost visibility for the most promising apps. Now, this analysis would require a preliminary understanding of the features that define a well-performing app. You can ask questions like:
- Does a higher size or price necessarily mean that an app would perform better than the other apps? 
- Or does a higher number of installs give a clear picture of which app would have a better rating than others?

### Data Handling and Cleaning

The first few steps involve making sure that there are no __missing values__ or __incorrect data types__ before we proceed to the analysis stage. These aforementioned problems are handled as follows:

 - For Missing Values: Some common techniques to treat this issue are
    - Dropping the rows containing the missing values
    - Imputing the missing values
    - Keep the missing values if they don't affect the analysis
 
    
 - Incorrect Data Types:
    - Clean certain values 
    - Clean and convert an entire column
 

```python
#checking the datatypes of all the columns of the dataframe
inp0.info()
```

#### Missing Value Treatment

```python
#checking the number of null values in the columns
inp0.isnull().sum()
```

Handling missing values for rating
 - Ratings is the target variable
 - drop the records

```python
#Drop the rows having null values in the Rating field
inp0 = inp0[~inp0.Rating.isnull()]

#checking the shape of the dataframe
inp0.shape
```

```python
# checking the number of nulls in the Rating field again to cross-verify
inp0.isnull().Rating.sum()
```

```python
#checking the number of nulls in the dataframe again and find the total number of null values
inp0.isnull().sum()
```

```python
#Inspect the nulls in the Android Version column
inp0[inp0['Android Ver'].isnull()]
```

```python
#Drop the row having shifted values
inp0 = inp0.drop(index = 10472)

#checking the nulls againin Android version column to cross-verify
inp0[inp0['Android Ver'].isnull()]
```

Imputing Missing Values

- For numerical variables use mean and median
- For categorical variables use mode

```python
#checking the most common value in the Android version column
inp0['Android Ver'].value_counts()
```

```python
#Fill up the nulls in the Android Version column with the mode value
inp0['Android Ver'] = inp0['Android Ver'].fillna(inp0['Android Ver'].mode()[0])
```

```python
#checking the nulls in the Android version column again to cross-verify
inp0['Android Ver'].isnull().sum()
```

```python
#checking the nulls in the entire dataframe again
inp0.isnull().sum()
```

```python
#checking the most common value in the Current version column
inp0['Current Ver'].value_counts()
```

```python
#Replace the nulls in the Current version column with the mode value
inp0['Current Ver'] = inp0['Current Ver'].fillna(inp0['Current Ver'].mode()[0])
```

#### Sanity Checks

The data that we have needs to make sense and therefore you can perform certain sanity checkings on them to ensure they are factually correct as well. Some sanity checkings can be:

- Rating is between 1 and 5 for all the apps.
- Number of Reviews is less than or equal to the number of Installs.
- Free Apps shouldn’t have a price greater than 0.


```python
#Perform the sanity checkings on the Reviews column
inp0['Reviews'].describe()
```

```python
# Check the cases where reviews are more than the installs
inp0[inp0.Reviews > inp0.Installs]
```

```python
# Drop these entries
inp0 = inp0[inp0.Reviews <= inp0.Installs]
# Check the cases where reviews are more than the installs again
inp0[inp0.Reviews > inp0.Installs]
```

```python
#perform the sanity checkings on prices of free apps 
inp0[(inp0.Price == 0.00) & (inp0.Type != 'Free')]
```

### Histograms

Histograms can also be used in conjuction with boxplots for data cleaning and data handling purposes. we can use it to check the spread of a numeric variable. Histograms generally work by bucketing the entire range of values that a particular variable takes to specific __bins__. After that, it uses vertical bars to denote the total number of records in a specific bin, which is also known as its __frequency__.


```python
#Create a histogram of the Reviews
plt.hist(inp0['Reviews'])
plt.show()
```

```python
#Create a boxplot of the Reviews column
plt.boxplot(inp0['Reviews'])
plt.show()
```

#### Distribution Plots

A distribution plot is pretty similar to the histogram functionality in matplotlib. Instead of a frequency plot, it plots an approximate probability density for that rating bucket. And the curve (or the __KDE__) that gets drawn over the distribution is the approximate probability density curve. 



```python
plt.hist(inp0.Rating)
plt.show()
```

```python
#Creating a distribution plot for rating
sns.displot(inp0.Rating, kde = True)
plt.show()
```

```python
#Changing the number of bins
sns.displot(inp0.Rating, kde = True, bins = 20)
plt.show()
```

```python
#Applying matplotlib functionalities
sns.displot(inp0.Rating, kde = True, bins = 20, color = 'lightgreen')
plt.title('Distribution of App Ratings', fontsize = 15)
plt.show()
```

**Pair Plots**

 - When you have several numeric variables, making multiple scatter plots becomes rather tedious. Therefore, a pair plot visualisation is preferred where all the scatter plots are in a single view in the form of a matrix
 - For the non-diagonal views, it plots a **scatter plot** between 2 numeric variables
 - For the diagonal views, it plots a **histogram**

```python
# Creating a pair plot for Reviews, Size, Price and Rating
sns.pairplot(inp0[['Reviews', 'Size', 'Price', 'Rating']])
plt.show()
```

```python
# Plotting a bar plot of Content Rating vs Average Rating 
inp0.groupby(['Content Rating'])['Rating'].mean().plot.bar()
```

```python
# Plotting the bar plot again with Median Rating
inp0.groupby(['Content Rating'])['Rating'].median().plot.bar()
```

```python
# Plotting the above bar plot using the estimator parameter
sns.barplot(inp0, x = 'Content Rating', y = 'Rating', estimator = np.median)
plt.show()
```

```python
# Plotting the bar plot with only the 5th percentile of Ratings
sns.barplot(inp0, x = 'Content Rating', y = 'Rating', estimator = lambda x: np.quantile(x, 0.05))
plt.show()
```

```python
# Plotting the bar plot with the minimum Rating
sns.barplot(inp0, x = 'Content Rating', y = 'Rating', estimator = np.min)
plt.show()
```

```python
# Plotting a box plot of Rating vs Content Rating
plt.figure(figsize=[10, 6])
sns.boxplot(x = inp0['Content Rating'], y = inp0['Rating'])
plt.show()
```

#### Heat Maps

Heat mapsutilise the concept of using colours and colour intensities to visualise a range of values. You must have seen heat maps in cricket or football broadcasts on television to denote the players’ areas of strength and weakness.

```python
## Ratings vs Size vs Content Rating

# Prepare buckets for the Size column using pd.qcut
inp0['Size_Bucket'] = pd.qcut(inp0.Size, [0, 0.2, 0.4, 0.6, 0.8, 1], ['VL', 'L', 'M', 'H', 'VH'])
```

```python
# Creating a pivot table for Size_buckets and Content Rating with values set to Rating
pd.pivot_table(inp0, index = 'Content Rating', columns = 'Size_Bucket', values = 'Rating')
```

```python
# Changing the aggregation to median
pd.pivot_table(inp0, index = 'Content Rating', columns = 'Size_Bucket', values = 'Rating', aggfunc = 'median')
```

```python
# Changing the aggregation to 20th percentile
pd.pivot_table(inp0, index = 'Content Rating', columns = 'Size_Bucket', values = 'Rating', aggfunc = lambda x: np.quantile(x, 0.2))
```

```python
# Storing the pivot table in a separate variable
res = pd.pivot_table(inp0, index = 'Content Rating', columns = 'Size_Bucket', values = 'Rating', aggfunc = lambda x: np.quantile(x, 0.2))
res
```

```python
# Plotting a heat map
sns.heatmap(res)
plt.show()
```

```python
# Applying customisations
sns.heatmap(res, cmap = 'Greens', annot = True)
plt.show()
```

```python
# Replacing Content Rating with Review_buckets in the above heat map
# Keeping the aggregation at minimum value for Rating
inp0['Review_Bucket'] = pd.qcut(inp0.Size, [0, 0.2, 0.4, 0.6, 0.8, 1], ['VL', 'L', 'M', 'H', 'VH'])
inp0.head()
```

```python
res1 = pd.pivot_table(inp0, index = 'Review_Bucket', columns = 'Size_Bucket', values = 'Rating', aggfunc = 'min')
res1
```

```python
# Applying customisations
sns.heatmap(res1, cmap = 'Greens', annot = True)
plt.show()
```

#### Stacked Bar Charts

- A stacked bar chart breaks down each bar of the bar chart on the basis of a different category
- For example, for the Campaign Response bar chart you saw earlier, the stacked bar chart is also showing the Gender bifurcation as well

```python
# Creating a pivot table for Content Rating and updated Month with the values set to Installs
pd.pivot_table(data = inp0, values = 'Installs', index = 'Updated Month', columns = 'Content Rating', aggfunc = 'sum')
```

```python
# Storing the table in a separate variable
monthly = pd.pivot_table(data = inp0, values = 'Installs', index = 'Updated Month', columns = 'Content Rating', aggfunc = 'sum')
monthly
```

```python
# Plotting the stacked bar chart.
monthly.plot(kind = 'bar', stacked = True, figsize = (10, 5))
plt.show()
```

```python
# Plotting the stacked bar chart again wrt to the proportions.
monthly_perc = monthly[['Everyone', 'Everyone 10+', 'Mature 17+', 'Teen']].apply(lambda x:x/x.sum(), axis = 1)
monthly_perc.plot(kind = 'bar', stacked = True, figsize = (10, 5))
plt.show()
```
